## 2. The Question: Decision, Action, Cost of a Wrong Call

### Research Question

Can machine learning estimate which search results are most relevant for a user's query so they can be ranked more effectively?

### Decision

The decision is determining which search results should appear higher in the search rankings.

### Who Acts?

The search ranking system or website administrator uses the predicted relevance scores to improve ranking.

### Action

Results predicted to be highly relevant can be ranked higher, while less relevant results can be ranked lower. This helps users find useful information more efficiently.

### Cost of a Wrong Recommendation

If an irrelevant page is ranked too highly, users may not find the information they need, reducing satisfaction and engagement. If a relevant page is ranked too low, useful content becomes harder to discover, potentially reducing website traffic and business value.

### Why Machine Learning?

Search relevance depends on many interacting features that are difficult to capture using manually written rules. Machine learning can learn these relationships from historical search data and provide better decision support.
